```python
# Load libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
df = pd.read_csv('../data/creditcard.csv')
df.head()
```

```python
# Check class distribution
df['Class'].value_counts()
```

```python
# Basic data info
df.info()
df.describe()
```

```python
# Visualize class imbalance
sns.countplot(data=df, x='Class')
plt.title('Class Distribution (0 = Legit, 1 = Fraud)')
```

```python
# Drop Time and Amount
filtered_df = df.drop(columns=["Time", "Amount"])

# Calculate correlation of all features with the Class column
correlation_matrix = filtered_df.corr()
correlation_with_class = correlation_matrix["Class"].drop("Class")  

# Convert to DataFrame for plotting
corr_df = correlation_with_class.sort_values().to_frame().reset_index()
corr_df.columns = ['Feature', 'Correlation']

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=corr_df, x='Correlation', y='Feature', palette='coolwarm')
plt.title("Correlation of Features with Fraud (Class = 1)")
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(), cmap='coolwarm', annot=False)
plt.title("Feature Correlation Heatmap")
```

```python
for col in ["V10", "V12", "V14", "V17"]:
    plt.figure(figsize=(6, 3))
    sns.boxplot(data=df, x='Class', y=col)
    plt.title(f"Distribution of {col} by Class")
    plt.show()
```

Key observations
- V12, V14, V17, and V10 show the strongest negative correlation with the fraud label. Fraudulent transactions tend to have lower values on those axes.
- The class-distribution plot confirms the extreme imbalance (~0.17 % fraud), which justifies the later use of SMOTE.
- Box-plots reinforce that the separation between classes is much clearer on V14 and V17 than on most other components.
- Features whose correlation hovers near zero carry less direct fraud signal but still help the model learn the legitimate class.

The bar chart ranks each PCA component by its correlation with the fraud label. Components V17, V14, V12, and V10 are the furthest from zero, all on the negative side, indicating that lower values on these axes are associated with fraudulent activity. During model training, Random Forest automatically gave higher importance to these features, which matched this exploratory finding. Components with correlation near zero provided background context for legitimate transactions but carried less direct fraud signal.
